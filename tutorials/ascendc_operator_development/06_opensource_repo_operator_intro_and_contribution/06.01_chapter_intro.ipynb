# CANN开源算子仓：昇腾NPU异构计算全场景算子开发、验证与贡献一站式开源平台
CANN开源算子仓不仅提供丰富的可参考算子资源，还为开发者搭建了便捷的算子贡献渠道。因仓库内算子需遵循统一规范进行适配使用，需在算子工程基础上添加指定格式的适配文件。本章将重点讲解各类适配文件的功能作用与编写方法，助力开发者快速完成算子参考与开源贡献。

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：
- 完成第一章学习，理解算子的核心概念与基本原理。
- 完成第二章学习，掌握基于Ascend C进行算子开发的基础方法。
- 成第三章进阶vector算子开发指导，了解算子工程的交付件结构与开发方式。

在环境配置方面，本章课程要求对你的环境满足以下条件：
- 系统中已部署昇腾NPU硬件，或已配置昇腾云服务器/仿真环境。
- 已按照[CANN下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。

---

## 章节目标
完成本课程后，你将能够：
- 熟悉开源算子仓的算子工程结构，实现算子资源的快速参考与使用
- 掌握开源算子仓的目录规范，能够高效完成算子的开源贡献
