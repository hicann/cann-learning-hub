# 融合算子开发
本章聚焦融合算子的开发实践，系统讲解融合算子的核心优势，并围绕 VV 融合、CV 融合两类典型算子的开发流程展开详细拆解，助力开发者精准掌握融合算子开发的核心要点与实操技巧。

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：
- 完成第一章学习，理解算子的核心概念与基本原理。
- 完成第二章学习，掌握基于Ascend C进行算子开发的基础方法。
- 完成第三章学习，掌握基于Ascend C进行算子工程开发的步骤，以及算子调用代码的编写。
- 完成第四章学习，掌握基于Ascend C进行Cube算子开发的要点。


在环境配置方面，本章课程要求对你的环境满足以下条件：
- 系统中已部署昇腾NPU硬件，或已配置昇腾云服务器/仿真环境。
- 已按照[CANN下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。

---

## 章节目标
完成本课程后，你将能够：
- 理解什么是融合算子，融合算子的核心优势。
- 掌握如何开发VV融合算子。
- 掌握如何开发CV融合算子。
