# Ascend C算子开发基础知识

本章将以矢量加法（Add）算子为例，系统介绍Ascend C算子核函数的完整开发与调用流程。内容涵盖Ascend C核心编程范式、核函数开发方法、主机侧调用实现以及最终交付件构成。通过本章学习，你将掌握使用Ascend C快速开发一个高性能矢量算子的全流程技能。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力:

- 完成第一章学习，理解算子的核心概念与基本原理。

在环境配置方面，本章课程要求对你的环境满足以下条件：

- 系统中已部署昇腾NPU硬件，或已配置昇腾云服务器/仿真环境。
- 已按照[CANN下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。

---

## 章节目标
完成本课程后，你将能够：

- 掌握核函数开发：编写符合Ascend C规范的核函数，理解其基本结构与语法。
- 掌握核函数调用：独立完成从主机侧调用核函数、管理数据与资源的完整流程。
- 理解编程范式：深入理解典型的Ascend C编程范式，特别是矢量编程范式的构建思想与核心概念（如TPipe, TQue, TBuf）。
- 运用API类库：了解并能查找使用Ascend C提供的各类API（如数据搬运、矢量计算、内存管理）。
- 实践算子构建：运用矢量编程范式，快速构建并验证一个完整的自定义算子。
