# Ascend C算子开发基础知识

本章围绕Ascend C算子开发的基础流程展开，先通过Hello World示例认识核函数编写与Host侧调用，再结合Add算子理解矢量编程范式、API使用、数据切分、流水任务和交付件构成。随后，本章进一步引入矩阵乘高阶API、融合算子概念、CV融合算子直调开发，以及PyTorch框架下通过pybind11和torch.library进行Kernel直调的方法。通过本章学习，你将掌握从设备侧核函数开发、Host侧验证到框架侧接入的基础实践路径。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力:

- 完成第一章学习，理解算子的核心概念与基本原理。

在环境配置方面，本章课程要求对你的环境满足以下条件：

- 系统中已部署昇腾NPU硬件，或已配置昇腾云服务器/仿真环境。
- 已按照[CANN下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。
- 如需学习PyTorch框架调用相关内容，需要额外完成PyTorch、torch_npu、pybind11等依赖的安装与配置。

---

## 章节目标
完成本课程后，你将能够：

- 掌握核函数开发：编写符合Ascend C规范的核函数，理解函数限定符、参数规则与设备侧入口结构。
- 掌握核函数调用：独立完成Host侧资源初始化、内存管理、核函数启动、结果回拷与验证流程。
- 理解编程范式：掌握典型Ascend C编程范式，重点理解矢量编程范式中的CopyIn、Compute、CopyOut、TPipe、TQue、TBuf和Double Buffer机制。
- 运用API类库：了解并能查找使用Ascend C提供的各类API（如数据搬运、矢量计算、内存管理）。
- 实践算子构建：基于Add算子完成核函数开发、CMake编译配置、运行验证与交付件理解。
- 理解融合算子：掌握融合算子的基本概念、性能收益，以及CV融合算子的典型执行流程。
- 完成框架接入：理解如何通过pybind11和torch.library将自定义Ascend C算子接入PyTorch调用链。
