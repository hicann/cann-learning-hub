# Asend C矩阵编程

本章将重点介绍Ascend C矩阵计算相关的编程知识，内容涵盖矩阵乘法基础，Ascend C矩阵乘相关分形格式,矩阵编程范式，使用高阶API开发矩阵乘的开发方法。通过本章学习，你将掌握使用Ascend C快速开发一个矩阵算子的全流程技能。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力:

- 完成第一章学习，理解算子的核心概念与基本原理。
- 完成第二章学习，理解Ascend C算子开发基础知识。
- 完成第三章学习，掌握如何使用算子工程方式开发算子。

在环境配置方面，本章课程要求对你的环境满足以下条件：

- 系统中已部署昇腾NPU硬件，或已配置昇腾云服务器/仿真环境。
- 已按照[CANN下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。

---

## 章节目标
完成本课程后，你将能够：
- 理解矩阵编程范式：深入理解Ascend C矩阵编程范式，特别是使用高阶API的矢量编程范式的构建思想。
- 掌握如何使用高阶API开发矩阵乘算子，了解并能查找使用Ascend C提供的各类高阶API实现矩阵乘法。
