# 基础概念入门：先把 GE 弄明白

<div align="left">
<blockquote style="margin-left: 0; margin-right: auto;">

欢迎来到 GE 图引擎开发教程的第一站。**快速入门**由两章组成——本章**基础概念入门**先帮你把 GE 的核心概念弄清楚，下一章**推理流程介绍**再带你动手把模型在昇腾上跑通。

本章聚焦「概念」：先讲清 GE 在昇腾全栈中的定位、两大职责、编译四阶段与模型下沉调度，再讲透用户构图所依赖的核心中间表示 AscendIR。这些概念是后续离线推理与在线执行实践的基础，无需任何昇腾 NPU 设备即可学习。

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：

- 能够阅读基础 C/C++ 代码，了解指针、缓冲区和资源管理的基本概念。
- 理解深度学习模型的基本概念，如算子、张量、计算图等。
- 了解 Host 内存、Device 内存的基本概念以及 Host-Device 数据搬运的流程。
- 了解 Linux 系统基本操作和命令行工具使用。
- 具备基础的 Python 编程能力。

无须具备任何 GE 图引擎相关的前置知识，也不要求提前掌握昇腾 NPU 底层的图编译和执行细节。本章内容**无需 NPU 设备**即可学习。

---

## 章节目标
完成本章后，你将能够：

- 了解 GE 在昇腾全栈中的定位，理解 GE「图编译」与「图执行」两大职责。
- 了解 GE 编译四阶段（图准备、图拆分、图优化、图编译）各自的工作内容与产物。
- 理解模型下沉调度机制，区分静态图整图下沉与动态 Shape 图的 Host 调度。
- 掌握 AscendIR 的核心概念，理解其对象模型及数据边/控制边/Anchor 的关系。
- 理解 Tensor 的 shape / dtype / format 描述，区分静态 Shape 图与动态 Shape 图。
- 理解 GE 的两条使用路径与框架驱动路径，能根据场景判断该用哪条。
