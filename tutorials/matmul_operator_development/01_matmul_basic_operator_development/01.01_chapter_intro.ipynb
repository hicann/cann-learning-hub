# 第一章：Matmul 算子基础开发

## 什么是 Matmul 算子

**Matmul（Matrix Multiplication，矩阵乘法）** 是线性代数中最核心的运算之一。给定两个矩阵 $A \in \mathbb{R}^{M \times K}$ 和 $B \in \mathbb{R}^{K \times N}$，它们的乘积 $C = A \times B$ 是一个 $M \times N$ 的矩阵，其中每个元素的计算方式为：

$$C_{i,j} = \sum_{k=1}^{K} A_{i,k} \cdot B_{k,j}$$

![Matmul 计算示意图](./images/mm_computaion.png)

在 AI 框架和算子开发领域，**Matmul 算子** 就是将这一数学运算封装为可在特定硬件上高效执行的算子（Operator）。在昇腾（Ascend）生态中，Matmul 算子基于 **Ascend C** 编程语言实现，充分利用昇腾 AI 处理器的 Cube 计算单元进行大规模并行矩阵运算。

除了标准的二维矩阵乘法之外，Matmul 算子通常还支持：
- **批量矩阵乘法（Batch Matmul）**：同时对多个矩阵对进行计算
- **混合精度计算**：支持 FP16/BF16/MXFP8/MXFP4 等多种数据类型
- **转置与广播**：支持输入矩阵的自动转置和维度广播

## 为什么 Matmul 算子如此重要

Matmul 算子是现代深度学习的**计算基石**，其重要性体现在以下几个方面：

1. **占据了神经网络的主要计算量**
   无论是经典的多层感知机（MLP）、卷积神经网络（CNN），还是当前主流的大语言模型（LLM）中的 Transformer 架构，矩阵乘法都是最频繁、最耗时的运算。以 Transformer 为例，自注意力机制（Self-Attention）中的 $QK^T$ 计算和前馈网络（FFN）中的全连接层，本质上都是 Matmul 运算。据统计，在大模型训练和推理过程中，**超过 90% 的浮点运算来自矩阵乘法**。

2. **直接影响模型训练和推理性能**
   Matmul 算子的执行效率直接决定了模型训练的速度和推理的延迟。一个高度优化的 Matmul 算子相比朴素实现，可以带来数倍甚至数十倍的性能提升。因此，深入理解 Matmul 算子的实现原理和优化技巧，是提升 AI 应用性能的关键。

3. **昇腾硬件架构的核心适配场景**
   昇腾 AI 处理器专为深度学习计算设计，其 Cube 单元专门用于加速矩阵乘法。学习和掌握 Ascend C 的 Matmul 算子开发，是发挥昇腾硬件全部算力的必经之路，也是 AI 开发者深入理解硬件-软件协同设计的绝佳切入点。

4. **算子开发能力的基础训练**
   Matmul 算子是学习算子开发的理想起点：通过 Matmul 算子开发，你可以系统地掌握数据搬运（Data Copy）、计算拆分（Tiling）、流式调度等 Ascend C 的核心开发技术，对 Cube 计算单元能有全面的认识与深入的理解，为开发其他更复杂的融合算子打下坚实的基础。
