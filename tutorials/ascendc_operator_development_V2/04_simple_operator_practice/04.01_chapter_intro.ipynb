# 简单算子实践

本章将在前面章节知识的基础上，通过多个完整算子案例介绍 Ascend C 算子的开发、调试与优化过程。课程覆盖基于指针的 C 语言编程、静态 Tensor、Tensor API、Cube-Vector 融合和 SIMT 编程等不同方式，帮助你将编程模型、数据切分、存储层级、流水同步和多核并行等概念应用到实际算子开发中。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力：

- 完成前三章学习，理解 Ascend C 算子的 Host 侧与 Device 侧开发流程，以及 SIMD、SIMT 编程模型的基本概念。
- 具备基础 C/C++ 编程能力，能够阅读指针、模板、Tensor 和简单的并行计算代码。
- 了解矩阵乘法、Softmax、Gather 等常见算子的基本功能，以及多核切分、数据搬运和同步等基础概念。

在环境配置方面，本章课程建议你的环境满足以下条件：

- 系统中已部署昇腾 NPU 硬件，或已配置课程所需的仿真环境。
- 已按照 [CANN 下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。
- 不同实践对硬件型号和运行模式的支持范围不同，运行前请先阅读对应课程中的环境与硬件说明。

---

## 章节目标

完成本章学习后，你将能够：

- 根据算子功能和输入输出规格，分析数据排布、数据切分和核函数实现方案。
- 使用基于指针的 C 语言编程、静态 Tensor、Tensor API 和 SIMT 等方式开发典型算子。
- 理解 Cube 与 Vector 协同、核间同步、多核并行和 double buffer 等优化方法的作用。
- 使用 CMake 和毕昇编译器构建 Ascend C 算子，并通过精度校验与性能分析检查实现结果。
- 根据算子的计算特点、访存模式和目标硬件能力，选择合适的编程方式与优化路径。
