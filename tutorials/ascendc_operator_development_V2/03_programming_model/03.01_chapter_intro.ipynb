# Ascend C 编程模型

本章将从 Ascend C 算子在 Host 侧与 Device 侧的运行关系出发，介绍 AI Core 的抽象硬件基础，以及面向 AI Core 的两类主要编程模型：SIMD 与 SIMT。通过本章学习，你将理解算子如何在多核、多线程和片上存储资源之间组织计算，并能够根据算子的数据访问模式、分支复杂度和目标硬件能力，初步判断应优先采用哪类编程模型。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力：

- 完成前两章学习，了解整个CANN的架构，理解算子的基本概念。
- 具备基础 C/C++ 编程能力，能够阅读函数定义、指针参数、模板函数和简单的内存访问代码。
- 了解并行计算的基本概念，例如多核并行、线程、数据分片和同步。

在环境配置方面，本章课程建议你的环境满足以下条件：

- 系统中已部署昇腾 NPU 硬件，或已配置昇腾云服务器/仿真环境。
- 已按照 [CANN 下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。
- 如果学习 SIMT 相关内容，需要确认目标硬件和软件版本支持 SIMT 编程能力。

---

## 章节目标

完成本章学习后，你将能够：

- 理解 Ascend C 算子在 Host 侧负责资源、数据和调度，在 Device 侧 AI Core 上执行计算的基本流程。
- 了解 AI Core 抽象硬件架构中的计算单元、存储单元和搬运单元，并理解数据在 Global Memory、Local Memory 和计算单元之间的流动方式。
- 掌握 SIMD 编程模型的核心思想，理解多核 SPMD 分片、单核 SIMD 批量计算、核函数定义与调用、以及常见 SIMD API 的学习路径。
- 掌握基于指针、基于 Tensor、基于 TPipe/TQue 的 SIMD 算子编程方式和使用场景。
- 掌握 SIMT 编程模型的基本结构，理解 Grid、Thread Block、Thread、线程索引、内存层级、同步机制和 SIMT API 的作用。
- 能够结合访存是否连续、数据形状是否规整、分支是否复杂、目标硬件是否支持等因素，初步判断算子更适合使用 SIMD 还是 SIMT。
