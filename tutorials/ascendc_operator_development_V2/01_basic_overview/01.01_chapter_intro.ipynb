# 第1章 前置知识 · 1.1 章节介绍
本系列教程面向 **昇腾950 系列**（950PR / 950DT，dav-3510 架构），讲解基于 Ascend C 的高性能算子开发。
950 搭载华为自研**第三代达芬奇（DaVinci）架构**，本系列引入其新特性，同时 **SIMD 部分跨代兼容 910B / 910C**。

---
## 昇腾950 系列简介
昇腾950 系列是面向下一代 AI 的旗舰计算芯片，采用多 Die 合封（2 个 AI Die + 2 个 IO Die + 片上内存模块），Chiplet UMA 统一内存架构：
- **昇腾950PR**：128GB / 1.6TB/s 片上内存，面向高性能推荐、大模型 Prefill 与多模态推理。
- **昇腾950DT**：144GB / 4TB/s 片上内存，面向大模型预训练、后训练与推理全流程。

## 950 关键特性总览
- **第三代 DaVinciCore（分离架构）**：36 个 AI 子系统，每个 = 1 个 **Cube Core** + 2 个 **Vector Core**。
- **Cube Core**：张量计算，原生支持 HiF8/MXFP8/FP8/MXFP4 低精度，MXFP4 峰值算力达上一代 4 倍。
- **Vector Core**：双发射 **Register-Based SIMD** 新架构，FP32/FP16 单核算力提升 100%，UB 与 ALU 间引入 RegFile 寄存器。
- **新同构 SIMD/SIMT 混合编程**：以 SIMD 为主、SIMT 为辅，面向 Gather/Scatter、复杂分支等离散场景。
- **NDDMA**：N 维直接内存访问引擎，最多 5 维数据重排与搬运。
- **CV 融合**：Cube L1 与 Vector UB 直接数据通路，加速 FlashAttention 等融合算子。
- **128MB 统一 L2 Cache** + **STARS2.0** 软硬协同调度。
- **灵衢互联 UB2.0**：单芯片互联带宽 2TB/s，超节点 8K 卡、集群超 128K 卡。
