## 1. 人工智能的发展历程
算子实际上是人工智能计算的“底层积木”，所以在了解算子之前，首先要对人工智能有一个基础的认知。
本节从 AI 的缘起出发，梳理其发展脉络与应用落地，帮助你建立宏观认知框架。


### 1.1 人工智能（AI）缘起：从图灵测试到达特茅斯会议
早在 1950 年，现代计算机科学之父阿兰・图灵在《计算机器与智能》中提出了**图灵测试**，首次将“机器是否具备人类智能”的探讨从哲学思辨推向科学研究维度。这一测试的核心逻辑是：测试者与被测试者（一人一机）相互隔离，通过键盘等装置随机提问；经过多次测试后，若机器能让平均每位测试者做出超过 30% 的误判，即被认为具备人类智能。


<img src="./images/turing_test.png">


> 上图展示了图灵测试的基本原理：测试者与被测试者（一人一机）相互隔离，通过键盘随机提问。若机器能让测试者做出超过 30% 的误判，即被认为具备人类智能。


值得一提的是，俄罗斯团队开发的聊天机器人尤金・古斯特，曾号称“史上首个通过图灵测试的智能体”（测试通过率 33%）。其核心思路是巧妙利用规则漏洞——冒充来自乌克兰、英语非母语的 13 岁小孩，在 5 分钟的测试时长内降低评委的判断标准，从而达成“骗过人类”的效果。
真正标志着人工智能成为独立研究学科的里程碑，是 1956 年的**达特茅斯会议**。约翰・麦卡锡、马文・明斯基、克劳德・香农等学者聚在一起，首次正式提出“人工智能”（Artificial Intelligence）概念，确立了用符号逻辑和计算方法模拟人类智能的研究范式，开启了 AI 的系统性研究。


### 1.2 AI 发展：从感知理解世界到生成创造世界
AI 的发展经历了几个关键阶段：
- **规则驱动阶段**（1950s–1970s）：基于人工编写的逻辑规则进行推理，代表系统如 ELIZA、DENDRAL。

- **深度学习阶段**（2010s–）：以 CNN、RNN、Transformer 为代表的深度神经网络在大数据与 GPU 算力加持下取得突破。
- **大模型阶段**（2020s–）：以 GPT、BERT 为代表的大规模预训练模型，具备跨领域多模态理解与生成能力。
AI 正从分析数据走向创造内容，从专用智能走向通用智能，大模型具备跨领域多模态能力，标志着从“感知理解世界”到“生成创造世界”的跃迁。

## 2. 什么是算子
在建立人工智能的基础认知后，我们将从算子含义和基本概念两方面，拆解算子的定义与核心构成。


### 2.1 算子的含义
人工智能的核心是模型算法，而算法本质是由一个个基础计算单元构成的，这些计算单元被称为**算子（Operator，简称 OP）**。在网络模型中，算子对应具体层或节点的计算逻辑，模型通常由一个或多个算子组合拼接而成。
如下图所示，基于 PyTorch 典型模型导出后，在 Netron 中可视化呈现的模型结构里，图中的 Conv 模块就是典型的卷积算子，而这类算子正是构成 AI 模型的基础计算单元。


<img src="./images/pytorch_net.png">


> 上图展示了 PyTorch 模型导出后在 Netron 中的可视化结构：图中的 Conv 模块即卷积算子，模型由多个这样的算子组合拼接而成，算子是构成 AI 模型的基础计算单元。


在数学领域，一个函数空间到函数空间上的映射 O：X→Y，都称为算子。广义地讲，对任何函数进行某一项操作都可以认为是一个算子，比如微分算子、不定积分算子等。在深度学习中，常见算子如 Conv2D（卷积）、MatMul（矩阵乘）、ReLU（激活）、Softmax、tanh、sigmoid 等。
算子是连接数学理论与硬件执行的桥梁：上接模型算法语义，下接芯片指令实现。Ascend C 算子开发的目标，就是让这座桥梁在昇腾 NPU 上跑得更快、更省、更灵活。

### 2.2 算子核心基本概念
理解算子需要掌握五个核心概念：**算子名称/类型、Tensor（数据容器）、TensorDesc（张量描述符）、shape（形状）、format（数据排布格式）**。下面逐一展开，并通过 NumPy 代码实操加深理解。


#### 2.2.1 算子名称（Name）与算子类型（Type）
- **算子名称（Name）**：网络中每个算子实例的唯一标识，类似变量名。如 Conv1、Conv2、MatMul_0。
- **算子类型（Type）**：决定算子的核心实现逻辑，同类型算子执行一致计算。如 Conv1 和 Conv2 的类型均为 Convolution。
区分名称与类型的意义在于：框架按类型调度算子实现，按名称管理数据流图。在 Ascend C 开发中，我们针对的是某一类型的算子实现，而非某个特定名称的算子实例。


<img src="./images/basic_concepts_of_operators_1.png">


> 上图展示了算子名称与算子类型的区别：Conv1、Conv2 是算子名称（唯一标识），它们的算子类型均为 Convolution（决定计算逻辑）。框架按类型调度实现，按名称管理数据流图。


#### 2.2.2 数据容器（Tensor）
**张量（Tensor）** 是承载算子输入/输出数据的多维数组容器，是标量（0 维）、向量（1 维）、矩阵（2 维）的高维拓展。Tensor 不关心数据含义，只负责按维度组织数值。在 CANN 中，Tensor 是算子计算的基本数据单元。
下面用 NumPy 的 ndarray 实操体验不同维度的 Tensor，并理解算子即计算单元的含义。


```python
import numpy as np

# 1. 创建不同维度的Tensor（NumPy的ndarray即为实操中的Tensor）
scalar = np.array(5)            # 0维Tensor（标量）
vec = np.array([1, 2, 3])       # 1维Tensor（向量）
mat = np.array([[1, 2], [3, 4]])# 2维Tensor（矩阵）
tensor_3d = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])  # 3维Tensor

print("0维:", scalar, "shape:", scalar.shape)
print("1维:", vec, "shape:", vec.shape)
print("2维:\n", mat, "shape:", mat.shape)
print("3维:\n", tensor_3d, "shape:", tensor_3d.shape)

# 2. 体验算子是基础计算单元：加法算子
add_op = lambda x, y: x + y
result = add_op(vec, vec)
print("\n加法算子 vec + vec =", result)
```

根据代码，尝试以下操作理解 Tensor 的概念。

- 尝试用减法/乘法定义新算子（如 mul_op = lambda x,y: x*y），对 Tensor 做计算。
- 故意让不同维度 Tensor 相加（如 add_op(vec, mat)），看报错，理解算子的计算约束。


#### 2.2.3 张量描述符（TensorDesc）
Tensor 作为数据容器，存放算子计算所需的实际输入/输出数值，相当于一杯装了水的杯子。

**张量描述符（TensorDesc）** 不存储任何实际数据，仅用于记录描述 Tensor 的元信息，相当于杯子上的标签。硬件层会先通过该标签读懂数据属性，再高效处理数据，这是适配异构计算、提升数据处理效率的经典设计。

TensorDesc 的核心属性包括：

| 属性 | 说明 |
|------|------|
| name | Tensor 的名称 |
| shape | 各维度的长度 |
| dtype | 数据类型（float16/float32/int8 等）|
| format | 数据排布格式（ND/NCHW/NHWC 等）|


```python
import numpy as np

def get_tensor_desc(tensor, name):
    return {"name": name, "shape": tensor.shape, "dtype": tensor.dtype,
            "format": "ND" if len(tensor.shape)>=1 else "SCALAR"}

img_tensor = np.random.rand(4,20,20,3).astype(np.float32)  # 4张20*20 RGB图片
desc = get_tensor_desc(img_tensor, "input_image")
print("TensorDesc:", desc)
print("\nTensor 存储实际数据，TensorDesc 只存元信息——标签+数据=完整描述")
```

- 修改 img_tensor 的 dtype 为 np.int8/np.bool_，看 TensorDesc 的 dtype 变化。


#### 2.2.4 形状（shape）
Tensor 的形状以 (D0, D1, ..., Dn-1) 的形式表示，D0 到 Dn 是任意正整数。如形状 (3,4) 表示第一维有 3 个元素，第二维有 4 个元素，是一个 3 行 4 列的矩阵。

典型 Tensor 与 shape 对应关系：

| 张量 | 形状 | 含义 |
|------|------|------|
| 5 | () | 标量 |
| [1,2,3] | (3,) | 一维向量 |
| [[1,2],[3,4]] | (2,2) | 二维矩阵 |
| shape=(4,20,20,3) | (4,20,20,3) | 4 张 20x20 RGB 图片 |

shape 可理解为操作 Tensor 的各层循环——每个维度对应一层 for 循环。


<img src="./images/detailed_explanation_of_shape.png">


> 上图展示了 shape 的含义：以 shape=(4,20,20,3) 为例，4 张 20x20 像素的 RGB 图片，每个维度对应一层 for 循环——shape 即循环层级。


```python
import numpy as np
img_np = np.random.randint(0,256,size=(4,20,20,3),dtype=np.uint8)
print("遍历第1张图前2行2列像素")
for i in range(1):
    for j in range(2):
        for p in range(2):
            for q in range(3):
                print(f"第{i+1}张-({j},{p})-{['R','G','B'][q]}:{img_np[i,j,p,q]}")
```

#### 2.2.5 数据排布格式（format）
卷积神经网络的特征图（Feature Map）通常用四维数组保存，即 4D 格式：
- **N**：Batch 数量，例如图像的数目。
- **H**：Height，特征图高度，即垂直高度方向的像素个数。
- **W**：Width，特征图宽度，即水平宽度方向的像素个数。
- **C**：Channels，特征图通道，例如彩色 RGB 图像的 Channels 为 3。

由于数据只能线性存储，因此这四个维度有对应的顺序。不同深度学习框架会按照不同的顺序存储特征图数据：
- Caffe / PyTorch：排列顺序为 [Batch, Channels, Height, Width]，即 **NCHW**。
- TensorFlow：排列顺序为 [Batch, Height, Width, Channels]，即 **NHWC**。

同一组数据在 NCHW 和 NHWC 下 shape 不同但语义相同，算子实现需感知 format 以正确寻址。


<img src="./images/data_layout_format_example.png">


> 上图展示了 NCHW 与 NHWC 两种数据排布格式的区别：同一组数据，Caffe/PyTorch 按 [N,C,H,W] 存储，TensorFlow 按 [N,H,W,C] 存储，算子需感知 format 以正确寻址。


```python
import numpy as np

# NCHW -> NHWC 转置示例
n, c, h, w = 1, 3, 4, 4
data_nchw = np.random.rand(n, c, h, w).astype(np.float32)
data_nhwc = np.transpose(data_nchw, (0, 2, 3, 1))  # NCHW -> NHWC

print("NCHW shape:", data_nchw.shape)
print("NHWC shape:", data_nhwc.shape)
print("数据相同，排布不同——算子需感知 format 正确寻址")
```

#### 2.2.6 轴（Axis）
Axis 是对 Tensor 维度的索引。对于一个 shape=(N,C,H,W) 的 4D Tensor：
- axis=0 对应 N（Batch）
- axis=1 对应 C（Channel）
- axis=2 对应 H（Height）
- axis=3 对应 W（Width）

很多算子按指定 Axis 执行计算，如 ReduceSum(x, axis=1) 表示沿通道维求和。理解 Axis 是编写维度相关算子（如 Concat、Split、Reduce）的基础。


```python
import numpy as np
x = np.ones((2, 3, 4, 5))  # NCHW
print("沿 axis=0 求和 (消除Batch维):", np.sum(x, axis=0).shape)
print("沿 axis=1 求和 (消除Channel维):", np.sum(x, axis=1).shape)
print("沿 axis=2 求和 (消除Height维):", np.sum(x, axis=2).shape)
```
