## Ascend C 多层级 API 总览


<img src="./images/architecture_ascendc.C76apSRf.png">

> 上图展示了 Ascend C 的 API 层次结构：从底到顶依次为 C API（`asc_*` 系列纯 C 接口）、Tensor C++ 编程（LocalTensor/GlobalTensor 抽象）、TPipe/TQue 框架（自动同步与内存管理）和 C++ 类库 API（基礎 API / 高阶 API），开发者可按需选择不同抽象层级。


Ascend C 提供多层级 API，均开放芯片完备编程能力，在开发效率与性能间取得平衡。从低到高依次为：

| API 层级 | 语言 | 编程范式 | 说明 | 本章使用 |
|---|---|---|---|---|
| 语言扩展层 C API（SIMD/SIMT） | C | 指针，数组 `[]` 分配 | `asc_*` 系列纯 C 接口，最底层 | ✅ 2.7 |
| Tensor C++ 编程 | C++ | Tensor 抽象，手动管理同步 | LocalTensor/GlobalTensor，自动布局转换 | ✅ 2.6 |
| TPipe/TQue 框架 API | C++ | Tensor，自动管理同步/内存 | 管道/队列管理，封装内存分配与同步 | ✅ 2.4 / 2.5 |
| C++ 类库 API · 基础 API | C++ | 连续计算 / 高维切分 | 封装常用矢量计算模式（Add、Mul 等） | ✅ 2.5 |
| C++ 类库 API · 高阶 API | C++ | 复用通用算法 | Matmul、Conv2d 等算子模板 | 否 |

> **选择原则**：追求极致性能与底层控制选 C API；追求开发效率选高阶 API；折中选 Tensor / TPipe-TQue。本章侧重入门，故选用 TPipe/TQue 框架和 Tensor 编程两个中间层级，覆盖 SIMD 与 SIMT 两种编程模型。


## 各节 API 选用与芯片兼容性
| 节 | 样例 | API 层级 | 编程模型 | 支持芯片 |
|---|---|---|---|---|
| 2.4 HelloWorld | hello_world | 核函数直调 | SIMD | Ascend 950PR/950DT、Atlas A3、Atlas A2 |
| 2.5 Add 算子 | add_custom | C++ 基础 API（TPipe/TQue） | SIMD | Ascend 950PR/950DT、Atlas A3、Atlas A2 |
| 2.6 Matmul 算子 | matmul_custom | C++ 基础 API（LoadData/Mmad） | SIMD | 仅 Ascend 950PR/950DT |
| 2.7 Gather 算子 | gather_1d_custom | SIMT（`blockIdx.x`/`threadIdx.x`） | SIMT | 仅 Ascend 950PR/950DT |

> **说明**：
> - 2.4/2.5 的 API 跨芯片兼容，编译时通过 `--npu-arch` 切换目标架构（`dav-3510` 对应 950，`dav-2201` 对应 A2/A3）。
> - 2.6 的矩阵算子涉及 Cube 矩阵布局转换，依赖 950 的 NDDMA 硬件特性，前代芯片不支持。
> - 2.7 的 SIMT 执行模式为 950 首次引入，编译需加 `--enable-simt` 选项，前代芯片不支持。
