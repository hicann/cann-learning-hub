# HIXL 的多种传输模式

本章聚焦于 HIXL 的 3 种传输模式（直传模式、中转 Buffer 模式以及 FabricMem 模式）的作用以及使用场景，帮助你掌握每种传输模式的优点和限制，以及能够根据实际的硬件环境和业务流量特点来选择合适的传输模式。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力：

- 完成第一章学习，理解单边通信的基本原理与 HIXL 的执行流程。
- 完成第二章学习，掌握基于 HIXL 核心 API 的使用方法。

---

## 章节目标

完成本课程后，你将能够：

- 理解直传模式、中转 Buffer 模式以及 FabricMem 模式的含义和原理。
- 了解每种传输模型的性能差异。
- 理解每种传输模式的适用场景。
