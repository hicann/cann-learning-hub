# HIXL 核心 API 使用方法

第 2 章介绍 HIXL C++ 核心 API 的生命周期，学习重点是理解每个接口在资源、内存、链路和传输四个阶段中的职责边界。HIXL 数据通路遵循固定顺序：Initialize → RegisterMem → Connect → Transfer → Disconnect → DeregisterMem → Finalize。

---


## 前置要求

为了充分掌握本章内容，你应已具备以下能力：

- 已完成第 1 章的学习，了解 HIXL 的基本概念和最小开发流程。
- 能阅读基础 C/C++ 代码，理解对象生命周期、指针和资源释放顺序。
- 了解 Linux 基本操作和命令行工具的使用。
- 具备基础内存管理概念，知道 Host/Device 内存申请和释放。

---


## 章节目标

完成本章学习后，你将能够：

- 解释 `Initialize` / `Finalize` 的参数、options、生命周期和释放顺序。
- 使用 `MemDesc`、`MemType`、`MemHandle` 描述并管理注册内存。
- 判断 `Connect` / `Disconnect` 的建链时机、超时设置。
- 区分 `TransferSync` / `TransferAsync`、`READ` / `WRITE` 和本地/远端地址语义。
