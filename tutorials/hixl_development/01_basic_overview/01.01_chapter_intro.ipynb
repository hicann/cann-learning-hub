# HIXL：面向昇腾 AI 场景的高性能单边通信库
为了让你轻松迈出 HIXL（Huawei Xfer Library）开发的第一步，本章会将所有核心概念进行介绍，包括单边通信与零拷贝基础、HIXL 架构、典型应用场景和最小开发流程等内容。这些概念将为你深入学习 HIXL 接口使用、链路建立、内存注册和数据传输奠定基础。

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：

- 能够阅读基础 C/C++ 代码，了解指针、缓冲区和资源释放顺序。
- 理解 Host 内存、Device 内存和设备间数据搬运的基本概念。
- 了解 Linux 系统基本操作和命令行工具使用。
- 具备基础的计算机网络知识。
- 熟悉大模型推理的流程。

无须具备任何 HIXL 和单边通信相关的前置知识，也不要求提前掌握昇腾 NPU 底层的通信链路细节。

---

## 章节目标
完成本课程后，你将能够：

- 理解单边通信、远端内存访问、零拷贝等通信术语。
- 说明 HIXL 的作用、整体架构、核心组件和典型应用场景。
- 掌握 HIXL 开发流程，理解一个最简实现样例的构成和执行流程。
