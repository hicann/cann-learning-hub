# PyAsc 认识与环境搭建
为了让你顺利迈出 PyAsc 算子开发的第一步，本章将开发前必须建立的核心认知一次性串起来：PyAsc 的项目定位与整体架构、工程目录与课程随附的算子样例、开发环境依赖与安装流程，以及如何端到端跑通第一个 PyAsc 样例。完成本章后，你将拥有一个可用的 PyAsc 开发环境，并能在 NPU 上板模式下跑通课程内提供的算子示例。

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：
- 能够看懂 Python 函数、类、装饰器等基础语法。
- 了解 Linux 系统基本操作（`cd`、`ls`、`pip` 等命令行工具）。
- 了解 Ascend C 算子开发的基本概念（推荐）：核函数、Global Memory、Local Memory、Tile 切分、流水同步等。如果不熟悉，可以先学习同仓库下的 [Ascend C 算子开发系列教程](../../ascendc_operator_development/01_basic_overview/01.01_chapter_intro.ipynb) 第 1 章。

无须具备任何 PyAsc 相关前置知识，也不要求提前掌握 PyAsc 编译器/MLIR 的实现细节。

在环境配置方面，本章课程要求对你的环境满足以下条件：
- 系统中已部署昇腾 NPU 硬件（Atlas A2 / A3），或已接入昇腾云开发环境（如 CANNLab）。
- 已完成 CANN 商用版/社区版（≥8.5.0.alpha001）的开发环境部署。
- Python 版本 3.9–3.12。

---

## 章节目标
完成本章学习后，你将能够：
- 说清 PyAsc 的项目定位、与 Ascend C 的关系，以及它在算子开发中扮演的角色。
- 描述 PyAsc 五大核心模块（编译运行、Python 前端、AST 转 ASC-IR、ASC-IR 定义、Ascend C 代码生成）的职责和协作方式。
- 看懂 PyAsc 工程目录结构，知道用户接口、后端实现、tutorials 样例分别位于哪里。
- 区分 PyAsc 的两种安装方式（pip 快速安装 vs 源码编译安装），并根据后续学习目标选择正确路径。
- 独立完成 CANN、PyAsc、运行时环境变量的配置，并在 NPU 上板模式下跑通课程提供的 Add 示例。
