# pyasc核函数开发基础

本章以矢量加法（Add）算子为主线，系统介绍pyasc核函数的完整开发与调用流程。内容涵盖pyasc编程范式、`@asc.jit`装饰器机制、Tensor与数据搬运、手动流水同步、核函数调用验证等核心知识。

### 学习前置要求

学习本章前，建议已经具备以下基础：

- 完成第1章学习，理解pyasc的基本概念和架构，完成环境搭建；
- 已学习Ascend C算子开发系列教程第2章，掌握核函数的基本概念和Ascend C编程范式；
- 了解AI Core的存储层级（Global Memory、Unified Buffer）和流水并行概念；
- 系统中已安装pyasc（v1.1.0及以上）和CANN软件（8.5.0及以上）。

### 学习目标

完成本章后，开发者应能够：

1. 理解pyasc的SPMD并行编程模型和Host-Device异构协同机制；
2. 掌握`@asc.jit`装饰器的使用和JIT编译流程；
3. 掌握`asc.GlobalTensor`和`asc.LocalTensor`的创建、管理和数据搬运；
4. 掌握手动流水同步（`asc.set_flag`/`asc.wait_flag`）机制和双缓冲编程；
5. 独立完成核函数的Launch调用和结果验证。
