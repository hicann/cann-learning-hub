# 3. 昇腾算子开发

本章将深入介绍SIMT算子开发和算子调试方法，帮助你掌握SIMT编程实践和调试技巧。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力：
- 已完成[基础掌握](../01_basic_mastery/01.00_chapter_intro.ipynb)章节的学习。
- 理解SIMT编程模型的基本概念。
- 熟悉核函数的定义与调用方法。

在环境配置方面，本章课程要求你的环境满足以下条件：
- 系统中已部署昇腾NPU硬件，或已配置昇腾云服务器/仿真环境。
- 已按照[CANN下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。

---

## 章节目标

完成本课程后，你将能够：
- 掌握SIMT算子开发的基本方法和实践技巧。
- 理解算子调试的常见方法和调试工具的使用。
- 能够独立开发并调试基础SIMT算子。
