# 第二章：加速库基础认知

本章介绍 CANN 算子库的标准目录结构和开发环境部署方法。以 FastGelu（标准 GELU 的 sigmoid 近似：`y = x / (1 + exp(-1.702 * x))`）算子为例，带领你理解算子工程的完整目录布局、各层文件的作用和关联，以及开发环境的安装与配置流程。

---

## 前置要求

为了充分掌握本章内容，你应:
- 已完成第一章（基础掌握）的学习
- 理解 AI Core 的基本架构和 regbase 编程模型

---

## 章节目标

完成本章后，你将能够：
- 理解算子的多种执行模式及其应用场景
- 掌握 aclnn 单算子 API 执行流程的详细步骤
- 理解 CANN 算子库的标准目录结构（op_host / op_kernel / op_api 等各层职责）
- 掌握 Tiling 的概念、参数组成和实现流程
- 了解 CMakeLists.txt 的编译配置和 build.sh 常用命令
- 能够安装 CANN Toolkit 和 Ops 包并配置开发环境
