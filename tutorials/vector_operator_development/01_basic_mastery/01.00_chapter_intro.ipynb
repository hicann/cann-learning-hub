# 1. 基础掌握

本章将系统介绍昇腾AI处理器的硬件架构、SIMD Membase编程基础和SIMT编程模型，为深入学习矢量算子开发奠定坚实基础。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力：
- 掌握C/C++基础编程能力，包括变量声明、循环、条件语句、指针与数组操作。
- 了解深度学习基础概念，熟悉常见AI算法的基本原理。
- 具备基础的计算机体系结构知识，了解并行计算概念。
- 已完成[Ascend C算子开发系列教程](../../ascendc_operator_development/01_basic_overview/01.01_chapter_intro.ipynb)，掌握Ascend C编程范式和核函数基本概念。

在环境配置方面，本章课程要求你的环境满足以下条件：
- 系统中已部署昇腾NPU硬件，或已配置昇腾云服务器/仿真环境。
- 已按照[CANN下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。

---

## 章节目标

完成本课程后，你将能够：
- 理解昇腾AI处理器A2/A3/A5系列芯片的架构特点，掌握AI Core内部结构。
- 掌握SIMD Membase编程范式的核心原理，理解DataBlock、Repeat、Mask等关键概念。
- 熟悉常见SIMD Membase指令的使用方法，能够编写基础矢量算子。
- 理解SIMT编程模型的抽象硬件架构、线程架构和内存层级。
- 掌握核函数的定义与调用，了解SIMT同步机制。
