# 第一章：基础掌握

本章介绍昇腾 AI 处理器的 Vector 计算架构，从芯片架构到 SIMD 指令集，再到 regbase 编程模型的基础知识，为后续的算子开发打下坚实根基。课程以 **FastGelu**（y = x / (1 + exp(-1.702 * x))）作为贯穿全课程的示例算子。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力：
- 具备 C++ 编程基础，能够编写函数、使用模板和指针
- 了解深度学习框架（PyTorch/TensorFlow）基本概念
- 了解计算机体系结构基础知识（CPU/GPU 内存模型、并行计算概念）
- 了解 Linux 系统基本操作及命令行工具的使用

在环境配置方面，本章课程要求你的环境满足以下条件：
- 系统中已部署昇腾 NPU 硬件（A2/A3/A5 系列），或已配置昇腾云服务器/仿真环境
- 已安装 CANN 开发环境（建议 CANN 8.0 以上版本）
- Python 3.8+ 环境，Jupyter Lab / Notebook

---

## 章节目标

完成本章后，你将能够：
- 理解昇腾 AI 处理器的 Vector 计算架构
- 了解 A2/A3/A5 芯片架构特点和 SIMD 指令集
- 掌握 SIMD Membase 编程范式的核心概念和常见指令
- 理解 Regbase 编程模型的核心概念和流水线架构
- 掌握 VF 寄存器的类型、操作流程和常用 API
- 了解 SIMT 编程模型的基础知识和线程架构
- 使用 VF 指令实现基本的向量计算（如 FastGelu、ReLU）
