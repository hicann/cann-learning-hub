# Ascend C：面向昇腾NPU的高性能算子开发语言
为了让你轻松迈出Ascend C算子编程的第一步，本章会将所有的核心概念进行介绍，包括算子基础原理、CANN异构计算架构概念、昇腾NPU架构、Ascend C简介等内容。这些概念将为你深入学习Ascend C算子编程奠定坚实的基础。

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：
- 能够在C++中声明变量、编写循环、并使用if/else条件语句。
- 理解指针、数组的概念及使用方法，能够通过指针访问数据。
- 掌握C/C++中函数定义、调用及参数传递的基本逻辑。
- 了解深度学习基础概念（如神经网络、张量、卷积等），熟悉常见AI算法的基本原理。
- 具备基础的计算机体系结构知识（如CPU/GPU内存模型、并行计算概念）。
- 了解Python基础语法及NumPy标准库的常规使用。
- 了解Linux系统基本操作及命令行工具的使用。

无须具备任何Ascend C相关前置知识，也不要求提前掌握昇腾NPU的底层架构细节。

在环境配置方面，本章课程要求对你的环境满足以下条件：
- 系统中已部署昇腾NPU硬件，或已配置昇腾云服务器/仿真环境。
- 已按照[CANN下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。

---

## 章节目标
完成本课程后，你将能够：
- 理解人工智能的发展历程、核心定义及行业应用场景。
- 掌握昇腾基础软硬件平台架构，包括 Atlas 系列产品、CANN异构计算架构的核心作用。
- 清晰认知算子的定义、基本概念（名称、类型、张量等）及在AI模型中的核心作用。
- 理解昇腾NPU的内部逻辑架构，尤其是AI Core的并行计算原理。
- 掌握Ascend C编程语言的核心特性及算子开发的方法。
