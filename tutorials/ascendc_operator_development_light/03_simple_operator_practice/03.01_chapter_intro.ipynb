# AscendC 算子实践开发

本章聚焦矩阵乘算子开发与CV融合算子开发实践，内容参考矩阵编程和融合算子开发章节，围绕矩阵乘法基础、Ascend C Matmul高阶API、融合算子概念、CV融合算子直调开发，以及Matmul和CV融合算子综合实践展开。通过本章学习，你将掌握使用Ascend C完成矩阵乘算子和典型融合算子的基础开发、编译运行与结果验证流程。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力:

- 完成第一章学习，理解算子的核心概念与基本原理。
- 完成第二章学习，理解Ascend C算子开发基础知识，掌握核函数编写、Host侧调用、Tiling设计和基础实践方法。
- 熟悉Ascend C工程的基本编译运行流程，能够阅读并补全 `.asc`、`CMakeLists.txt` 等示例代码。

在环境配置方面，本章课程要求对你的环境满足以下条件：

- 系统中已部署昇腾NPU硬件，或已配置昇腾云服务器/仿真环境。
- 已按照[CANN下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。
- 如需运行直调样例，需要正确配置Ascend C编译器、CMake和运行时环境变量。

---

## 章节目标
完成本课程后，你将能够：

- 理解矩阵乘法基础：掌握矩阵乘的数据形状、数据格式、分块思路和矩阵编程范式。
- 掌握Matmul高阶API开发：能够基于Ascend C Matmul高阶API完成矩阵乘算子的Kernel实现、Host侧Tiling生成和直调验证。
- 理解融合算子：掌握融合算子的基本概念、性能收益和典型执行流程。
- 掌握CV融合算子开发：能够完成Matmul与Vector后处理融合的直调工程开发，并理解数据流、Tiling和Kernel实现之间的关系。
- 完成综合实践：能够独立补全Matmul和CV融合算子练习代码，编译运行并验证输出结果。
