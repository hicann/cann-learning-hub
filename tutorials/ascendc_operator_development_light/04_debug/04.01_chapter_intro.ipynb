# Ascend C算子调试调优

本章聚焦Ascend C算子开发过程中的调试调优能力，围绕算子调试、算子上板调优和算子仿真调优三个典型场景展开。通过本章学习，你将掌握如何在算子开发验证阶段观察运行时数据、定位日志信息、采集性能指标，并结合上板和仿真性能数据分析性能瓶颈。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力：

- 已完成Ascend C基础课程学习，理解核函数、Host侧调用、CMake编译和Add算子样例的基本结构。
- 了解NPU上板执行与仿真执行的差异，能够阅读基础的Ascend C Kernel代码。

在环境配置方面，本章课程要求你的环境满足以下条件：

- 已完成CANN开发环境部署，并可正常使用Ascend C编译工具链。
- 如需执行算子调试和算子上板调优课程，需要具备可用的昇腾NPU硬件环境。
- 如需执行算子仿真调优课程，需要具备Ascend C仿真编译与msProf simulator相关工具环境。

---

## 章节目标
完成本章后，你将能够：

- 掌握算子调试：使用DumpTensor和printf输出运行时数据与日志，辅助定位Kernel侧逻辑问题。
- 掌握算子上板调优：使用msProf采集单算子程序的上板性能数据，并阅读关键性能文件。
- 掌握算子仿真调优：使用msProf simulator采集仿真trace，结合源码行信息定位性能热点。
- 完成针对性优化：通过屏蔽调试打印等方式对比优化前后的性能数据，理解调试信息对执行耗时的影响。
