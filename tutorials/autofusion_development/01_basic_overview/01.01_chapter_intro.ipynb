# AutoFuse 基础介绍

本章帮助你建立 AutoFuse 的整体认知并掌握基础使能方式，从"什么是 AutoFuse"开始，逐步介绍其核心价值、接入路线、融合过程及开启方式等。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力：

- 熟悉 CANN、昇腾 NPU 的基础概念，了解深度学习模型由算子组成的基本结构。
- 理解 Tensor、shape、dtype 等深度学习核心概念。
- 了解昇腾环境中模型经图编译后执行的基本过程。

---

## 章节目标

完成本章后，你将能够：

- 说明 AutoFuse 解决的核心问题，理解其缓解 Memory Bound 场景的原理。
- 区分 AutoFuse 与业务代码直接调用 API 的差异，理解其在图编译流程中的作用。
- 了解 AutoFuse 的两条接入路线及融合过程的整体脉络。
- 掌握 `AUTOFUSE_FLAGS` 配置方法，完成 AutoFuse 基础使能。
