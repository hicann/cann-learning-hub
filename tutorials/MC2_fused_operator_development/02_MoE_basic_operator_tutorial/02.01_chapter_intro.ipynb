# MoE 算子开发教程


## 课程简介

本教程是昇腾 CANN（Compute Architecture for Neural Networks）算子开发系列教程的高级篇章，聚焦于大模型领域最前沿的 **MoE（Mixture of Experts，混合专家模型）** 架构下的通信算子开发。课程从 MoE 架构原理出发，逐步深入到 Dispatch/Combine 核心算子的实现细节，并通过量化 Dispatch 实战项目帮助开发者掌握高性能 MoE 通信算子的开发技能。

MoE 架构通过稀疏激活机制，在保持模型规模增长的同时显著降低实际计算量，已成为超大规模模型的关键技术路线。本教程将帮助开发者深入理解 MoE 算子的设计思想与优化技巧，为构建高效的分布式大模型推理系统打下坚实基础。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力：

- 完成ascendc_operator_development第一章学习，理解算子的核心概念与基本原理。
- 完成ascendc_operator_development第二章学习，掌握基于Ascend C进行算子开发的的基础方法。
- 完成ascendc_operator_development第三章学习，掌握基于Ascend C进行算子工程开发的的步骤，以及算子调用代码的编写。
- 了解分布式训练基本概念，理解集合通信操作（如AlltoAll、AllReduce等）。

---

## 课程目标

学完本教程后，您将能够：

- **理解 MoE 架构原理**：掌握 Mixture of Experts 的核心概念、门控网络、专家网络及 Top-K 路由策略
- **掌握并行策略设计**：深入理解专家并行（EP）、张量并行（TP）等 MoE 并行方案
- **开发 Dispatch/Combine 算子**：掌握 MoE 核心通信算子的接口设计、数据流处理与跨卡通信机制
- **实现高性能优化**：学习 Win 区内存布局、Tiling 策略、流水并行等关键优化技术
- **完成量化实战项目**：动手实现支持 MXFP8 动态量化的高性能 Dispatch 算子

## 实践项目

本教程提供 **量化 Dispatch 算子实践项目**，位于 [`scaffold_02_10/`](./scaffold_02_10/) 目录：

- **项目目标**：实现支持 MXFP8 动态量化的 Dispatch 算子，完成 Token 分发、跨卡通信与数据整理
- **技术要点**：量化计算、共享内存通信、流水并行、多核同步
- **项目结构**：包含完整的 CMake 构建系统、头文件、源码框架及测试脚本
- **实践方式**：补全 `include/moe_distribute_dispatch.h` 中 3 处核心函数实现


建议按照章节顺序学习，理论章节（02.02-02.06）帮助建立完整知识体系，优化章节（02.07-02.09）深入理解MOE算子，实践章节（02.10）将所学知识应用于实战项目。
