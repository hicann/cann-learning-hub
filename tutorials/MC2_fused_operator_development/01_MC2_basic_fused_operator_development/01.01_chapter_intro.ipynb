# MC2融合算子开发
本章聚焦MC2(Matrix Computation & Communication)融合算子的开发实践，系统讲解计算与通信融合的核心优势，并围绕Matmul+MTE通信典型算子的开发流程展开详细拆解，助力开发者精准掌握MC2融合算子开发的核心要点与实操技巧。

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：
- 完成ascendc_operator_development第一章学习，理解算子的核心概念与基本原理。
- 完成ascendc_operator_development第二章学习，掌握基于Ascend C进行算子开发的基础方法。
- 完成ascendc_operator_development第三章学习，掌握基于Ascend C进行算子工程开发的步骤，以及算子调用代码的编写。
- 完成ascendc_operator_development第四章学习，掌握基于Ascend C进行Cube算子开发的要点。
- 完成ascendc_operator_development第五章学习，掌握VV融合算子和CV融合算子的开发方法。
- 了解分布式训练基本概念，理解集合通信操作（如AllReduce、ReduceScatter等）。

在环境配置方面，本章课程要求对你的环境满足以下条件：
- 系统中已部署多卡昇腾NPU硬件环境（至少2卡），或已配置多卡昇腾云服务器/仿真环境。
- 已按照[CANN下载页面](https://www.hiascend.com/cann/download)完成对应硬件的开发环境部署。
- 已配置HCCL集合通信库环境。
