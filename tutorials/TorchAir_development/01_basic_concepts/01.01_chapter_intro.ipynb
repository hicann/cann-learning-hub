# TorchAir：昇腾torch_npu的图模式能力扩展库
TorchAir（Torch Ascend Intermediate Representation）是昇腾Ascend Extension for PyTorch（torch_npu）的图模式能力扩展库，提供了昇腾设备亲和的torch.compile图模式后端，实现了PyTorch网络在昇腾NPU上的图模式推理加速以及性能优化。

---

## 前置要求

在开始本教程之前，建议你具备以下基础知识与技能：
- 熟悉 Python 语言，以及 PyTorch 深度学习框架的基本用法。
- 了解 `torch.compile` 机制与 FX 计算图（FX Graph）的基本概念。
- 了解昇腾 NPU 硬件加速与 `torch_npu` 驱动的基本概念。

---

## 章节目标
完成本课程后，你将能够：

- 掌握 PyTorch `torch.compile` 整体运行流程、核心参数作用；
- 理解 NPUGraph **Capture 捕获 + Replay 回放** 双阶段核心原理；
- 清晰掌握`npugraph_ex`后端定位、专属基础概念，熟悉该后端的适用场景与硬件等；
- 建立从`torch.compile`上层接口 → NPUGraph 框架 → `npugraph_ex`昇腾专用后端完整知识链路，为后续实操、Profiling 性能对比实验铺垫理论基础。
