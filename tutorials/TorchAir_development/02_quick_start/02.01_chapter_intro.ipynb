# 快速入门
本章为TorchAir图模式实操入门章节，承接上一章理论原理内容，章节从环境校验起步，提供标准化 SimpleModel 示例工程，完整演示`npugraph_ex` 图模式代码的编写与运行流程。

---

## 前置要求

在开始本教程之前，建议你具备以下硬件环境：
- 确认有可用的NPU环境，推荐使用[CANNLab](https://devstation.connect.huaweicloud.com/devdesktop/webide)一站式开发平台
- CANN 版本、torch 与 torch_npu 版本配套，[版本配套关系参考](https://gitcode.com/Ascend/pytorch#ascend-extension-for-pytorch%E7%89%88%E6%9C%AC%E9%85%8D%E5%A5%97%E8%A1%A8)
---

## 章节目标
完成本课程后，你将能够：

- 能够独立完成昇腾NPU环境校验，熟练编写适配`npugraph_ex`后端的完整推理代码，掌握torch.compile关键参数配置规范；
- 区分原生 `Eager` 逐算子执行模式与 `npugraph_ex` 图模式的代码写法，理解两种模式在执行流程、推理时延上的直观差异；
- 掌握昇腾 Profiler 性能采集代码编写方法，可分别采集 `Eager`、`npugraph_ex` 两种模式的性能 Trace 文件，并分析两者的性能收益。
