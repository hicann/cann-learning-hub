## 从 SFT 到 RL：思维方式的转变

在 SFT 中，我们告诉模型「这样做是对的」——通过标准答案和 cross-entropy loss 直接纠正模型的输出。

在 RL 中，我们告诉模型「这样做得了多少分」——通过奖励函数评估模型的输出，让模型自己发现什么是好策略。

这个转变看似简单，但带来了几个根本性的变化：

1. **没有标准答案**：RL 只需要定义「什么是好的」，而不需要知道「最好的是什么」。
2. **需要探索**：模型必须尝试不同的策略才能发现更好的方法，这就引入了探索与利用的平衡。
3. **多步决策**：在 Wordle 中，每一步猜词都会影响后续的反馈和决策，模型需要学会长远规划。

理解这些差异，是掌握后续所有 RL 概念的起点。

```text
SFT: 告诉模型正确答案
  loss = CrossEntropy(model_output, correct_answer)
  -> 模型学习模仿正确答案

RL: 告诉模型得分
  reward = reward_function(model_output)
  -> 模型学习获得更高分数

  关键问题：模型如何知道哪些输出得分高?
  -> 生成多条输出，比较它们的得分（这就是 GRPO 的核心思想）
```
