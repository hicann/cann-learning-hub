# Wordle RL 训练实战

在上一章中，我们学习了 RL 的核心概念：策略、奖励、优势、PPO、GRPO、KL 散度和训练稳定性。本章将把这些概念付诸实践——运行 Wordle GRPO 训练，观察真实的训练过程，解读训练指标。

Wordle 是一个经典的猜词游戏，非常适合作为 RL 入门实战：规则简单、多轮交互、奖励可量化。通过本章，你将看到 RL 理论如何转化为实际的训练命令和训练曲线。

> 本教程的 Wordle 训练样例参考了 [prime-rl](https://github.com/PrimeIntellect-ai/prime-rl/tree/main/examples/wordle) 项目的 Wordle 示例，包括游戏环境、奖励函数设计和数据生成方法。
