# 训练调优与问题排查

在第 3 章中，我们成功运行了 Wordle GRPO 训练。但实际训练中，你可能会遇到各种问题——训练不收敛、策略崩塌、entropy 失控等。本章将基于真实的训练实验数据，介绍如何调优超参数和排查训练问题。

本章的内容全部来自真实的训练实验数据，而非理论推测。你将看到同一模型在不同超参数下的表现差异，以及如何通过训练指标诊断和解决问题。

---

## 前置要求

为了充分掌握本章内容，你应已具备以下能力：

- 完成第 1-3 章学习，理解 RL 核心概念并成功运行过 Wordle 训练。
- 理解 entropy、KL 散度、PPO clip 等概念。
- 能够解读训练日志中的关键指标。

---

## 章节目标

完成本章后，你将能够：

- 理解 entropy_coeff、kl_loss_coef、学习率等超参数对训练的影响。
- 诊断策略崩塌和 entropy 失控等常见训练问题。
- 根据训练指标选择合适的超参数组合。
