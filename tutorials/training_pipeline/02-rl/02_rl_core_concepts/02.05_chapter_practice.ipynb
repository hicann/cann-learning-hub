## 选择题

1. （判断题）策略（Policy）就是语言模型本身，给定输入后输出一个 token 概率分布。

2. （判断题）参考策略（π_ref）在每个训练 step 都会更新，与当前策略保持同步。

3. （判断题）GRPO 不需要 Critic 模型，用组内平均奖励作为 baseline 来计算优势。

4. （判断题）PPO 的 clip 机制限制每次策略更新的幅度，防止策略突变。

5. （判断题）Entropy bonus 设置得越大越好，可以充分保证模型的探索能力。

6. （判断题）KL 散度正则化通过在 loss 中加入惩罚项，限制当前策略与参考策略之间的距离。
