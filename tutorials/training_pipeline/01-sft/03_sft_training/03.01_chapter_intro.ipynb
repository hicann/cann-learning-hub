## 预期成果

根据附录 A 的参考结果，20 步 SFT 训练后预期：

| 指标 | 训练前 | 训练后 | 说明 |
|------|--------|--------|------|
| `format_reward` | ~0.200 | ~1.000 | SFT 核心目标：格式对齐 |
| `avg reward` | ~0.04 | ~0.40 | 综合得分提升 10× |
| `loss` | ~1.25 | ~0.49 | CE Loss 快速下降 |
| `correct_answer` | 0% | 0% | 策略留给 RL 阶段 |

> **核心预期**：SFT 后模型 100% 遵循 `<guess>word</guess>` 输出格式，稳定完成多轮对话。猜词策略的提升是 RL 阶段的任务。
