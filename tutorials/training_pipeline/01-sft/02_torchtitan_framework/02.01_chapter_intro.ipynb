# TorchTitan 框架概述与环境配置

第 1 章我们理解了 SFT 的概念和 Wordle 任务。现在需要回答一个实际问题：**SFT 训练怎么跑起来？**

答案是用 **TorchTitan**——一个基于 PyTorch FSDP 的模块化大模型训练框架。本章从 FSDP 原理出发，理解 TorchTitan 如何把分布式训练包装成一条命令启动，最后动手配好环境。
