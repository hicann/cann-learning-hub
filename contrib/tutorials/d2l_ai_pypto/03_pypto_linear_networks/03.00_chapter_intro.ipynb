# pypto 线性神经网络
本章内容改编自[《Dive into Deep Learning》第3章「线性神经网络」](https://zh.d2l.ai/chapter_linear-networks)，使用 pypto 框架在昇腾 NPU 上实现。涵盖线性回归、softmax 回归及其在图像分类中的应用。

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：
- 掌握 Python 基础语法，能够编写和运行 Python 脚本。
- 了解张量的基本操作（创建、运算、索引及广播）。
- 理解线性代数、微积分、自动微分的基础知识。

---

## 章节目标
完成本课程后，你将能够：
- 理解线性回归的基本原理和实现方式。
- 掌握 softmax 回归在多分类问题中的应用。
- 使用 pypto 在 NPU 上构建、训练和评估线性模型。
- 理解图像分类数据集的处理方法。
