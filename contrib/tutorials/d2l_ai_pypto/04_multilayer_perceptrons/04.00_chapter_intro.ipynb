# pypto 多层感知机
本章内容改编自[《Dive into Deep Learning》第4章「多层感知机」](https://zh.d2l.ai/chapter_multilayer-perceptrons/)，使用 pypto 框架在昇腾 NPU 上实现。涵盖多层感知机、过拟合与欠拟合、正则化技术（权重衰减、暂退法）、数值稳定性与参数初始化等关键主题。 

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：
- 掌握 Python 基础语法，能够编写和运行 Python 脚本。
- 了解张量的基本操作（创建、运算、索引及广播）。
- 理解线性回归和 softmax 回归的基本原理。
- 了解自动微分和反向传播的基本概念。

---

## 章节目标
完成本课程后，你将能够：
- 理解多层感知机的基本架构和原理。
- 掌握过拟合、欠拟合的概念及模型选择方法。
- 使用权重衰减和暂退法等正则化技术防止过拟合。
- 理解前向传播、反向传播的计算过程。
- 掌握数值稳定性和参数初始化的关键技巧。
- 将所学知识应用于真实的房价预测案例。
