# pypto 预备知识
本章内容改编自[《Dive into Deep Learning》第2章「预备知识」](https://zh.d2l.ai/chapter_preliminaries)，使用 pypto 框架在昇腾 NPU 上实现。涵盖数据操作、数据预处理、线性代数、微积分、自动微分和概率等深度学习所需的基础知识与工具。

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：
- 掌握 Python 基础语法，能够编写和运行 Python 脚本。
- 了解基本的数学概念，如标量、向量、矩阵等。

无须具备深度学习相关前置知识，也不要求提前掌握昇腾 NPU 的底层架构细节。

在环境配置方面，本章课程要求满足以下条件：
- 拥有 GitCode 账号，能够使用 CANNLab 创建 NPU 开发环境，使用 GitCode notebook。

---

## 章节目标
完成本课程后，你将能够：
- 使用 pypto 在 NPU 上进行张量的创建、运算、索引及广播等基本操作。
- 使用 pandas 对原始数据进行预处理，并将其转换为张量格式。
- 理解线性代数、微积分、概率等数学基础知识在深度学习中的应用。
- 使用自动微分工具计算梯度。
