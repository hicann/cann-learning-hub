# pypto 快速开始
本章将帮助你快速上手 pypto 框架，包含使用 CANNLab 创建 NPU 开发环境、常用环境持久化配置、使用 notebook 进行 pypto 开发等内容。

---

## 前置要求
为了充分掌握本章内容，你应已具备以下能力：
- 掌握 Python 基础语法，能够编写和运行 Python 脚本。
- 了解 Linux 系统基本操作及命令行工具的使用。

无须具备任何 pypto 相关前置知识，也不要求提前掌握昇腾 NPU 的底层架构细节。

在环境配置方面，本章课程要求满足以下条件：
- 拥有 GitCode 账号，能够使用 CANNLab 创建 NPU 开发环境，使用 GitCode notebook。

---

## 章节目标
完成本课程后，你将能够：
- 使用 CANNLab 快速开发 pypto。
- 在 CANNLab 上进行常用环境的持久化配置。
- 使用 notebook 快速开发 pypto。
